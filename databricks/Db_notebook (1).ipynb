###Create Catalog


```python
%sql
CREATE CATALOG sales_catalog 
MANAGED LOCATION 'abfss://metastore@dlsaleslakehouse.dfs.core.windows.net/' ;
```

###Create Schema

```python
%sql
CREATE SCHEMA sales_catalog.silver
```

```python
%sql
CREATE SCHEMA sales_catalog.gold
```

```python
%sql
DROP TABLE IF EXISTS sales_catalog.silver.silver_watermark;
CREATE TABLE IF NOT EXISTS sales_catalog.silver.silver_watermark (
    table_name STRING,
    last_watermark STRING,
    updated_at TIMESTAMP
)
USING Delta
LOCATION 'abfss://silver@dlsaleslakehouse.dfs.core.windows.net/watermark'
```

```python
%sql
INSERT INTO sales_catalog.silver.silver_watermark VALUES  ('sales', 'DT00000', current_timestamp)
```
